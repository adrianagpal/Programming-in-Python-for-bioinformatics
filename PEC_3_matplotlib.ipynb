En este ejercicio tendremos que crear dos funciones en que las suyas características tienen que ser las siguientes:
* **es_digit()**: Creáis una función en que la variable de entrada se puede corresponder a un texto en que hay letras y números y tiene que devolver solo las letras en una sola variable en formato de string.
* **clasifica_classes_CAZymes()**: Creáis una función en que la variable de entrada sea un fichero de texto descargado de la base de datos CAZy y que clasifique en un diccionario, las clases de CAZymes que hay en aquel fichero y sus valores se corresponden a la cantidad existente por cada clase.

Finalmente, feudo un gráfico de diagrama de sectores (o pie chart en ángulos) en que se vean los porcentajes para cada clase, tal y como se ve en la siguiente figura.


image.png

````python
# para poder visualizar el gráàfico en este notebook
%matplotlib inline

# importamos los módulos necessarios
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Función que encuentra si un valor de un string es un digito o no, y solamente develve
# los valores correspondientes a las letras
def es_digit(data):

    # Recorremos la string
    for caracter in data:

        # Extraemos las letras en forma de string
        letras = "".join(ch for ch in data if ch.isalpha())

    return letras

# Función en que creamos un diccionario en que las keys son el tipo de clases
# que pueden llegar a existir por los CAZymes y los valores se corresponde a la
# cantidad existente de estas clases.
def clasifica_classes_CAZymes(data):

  # Creamos variable donde guardar nuestro diccionario
  CAZymes = {}

  # Abrimos el fichero en modo lectura
  with open(data, "r") as infile:

    # Recorremos el fichero línea por línea
    for line in infile:

        # Extraemos nombre enzima del primer elemento eliminando los dígitos
        enzima = es_digit(line.split()[0])

        # Si la enzima se encuentra en las claves del diccionario, sumamos 1
        if enzima in CAZymes.keys():
            CAZymes[enzima] += 1

        # En caso contrario, creamos clave con valor 1
        else:
            CAZymes[enzima] = 1

    # Devolvemos resultado del diccionario
    return CAZymes

# Guardamos valor del diccionario en la variable
dic_enzimas = clasifica_classes_CAZymes('cazy_data_Aspergillus.txt')

# Creamos figura
fig, ax = plt.subplots()

# Creamos gráfico con los valores del diccionario en porcentaje
ax.pie(dic_enzimas.values(), autopct='%1.1f%%')

# Añadimos leyenda con claves del diccionario y título
plt.legend(dic_enzimas.keys(), loc="upper right")
plt.title("Clasificación CAZymes d'Aspergillus")

# Mostramos gráfico
plt.show()
````

### Ejercicio 2 **<font color="blue" size="+2">EG</font>**

Representa en un gráfico 3D la función `f(x,y) = 4*cos(x) + 3*sin(y)` en el intervalo `[-2(pi/2), 2(pi/2)]` (en radianes) con un step de `0.1`

Podéis consultar ejemplos de graficos 3D aquí: https://matplotlib.org/stable/plot_types/index.html

````python
# La representación debe dar una imagen como la siguiente
````

image.png

````python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

plt.rcParams['figure.figsize'] = 4.5, 4.5

# Respuesta

# Definimos la función que queremos representar
def fun(x, y):
    return 4*np.cos(x) + 3*np.sin(y)

# Creamos una nueva figura
fig = plt.figure()

# Añadimos un subgráfico tridimensional
ax = fig.add_subplot(111, projection='3d')

# Definimos vectores para los ejes x e y
x = y = np.arange(-np.pi, np.pi, 0.1)

# Creamos matrices de coordenadas a partir de los vectores de los ejes
X, Y = np.meshgrid(x, y)

# Creamos un array a partir de la función de ambas matrices
Z = np.array(fun(X, Y))

# Añadimos el gráfico a la figura
ax.plot_surface(X, Y, Z, cmap = cm.coolwarm, alpha = 1)

# Mostramos el gráfico
plt.show()
````

### Ejercicio 3 **<font color="blue" size="+2">EG</font>**
Representa la función que quieras de la forma que consideres. Sé creativo a la hora de escoger la función a representar y el intervalo de valores.

Podéis consultar https://matplotlib.org/stable/plot_types/index.html para inspiraros.

````python
%matplotlib inline

# Respuesta

import matplotlib.pyplot as plt
import numpy as np

# Definimos función para la representación de toroides
def repr_torus(definicion, param1, param2):

    # Definimos arrays con intervalo de valores
    theta = np.linspace(0, 2*np.pi, definicion)
    phi = np.linspace(0, 2*np.pi, definicion)

    # Combina los arrays en una matriz
    theta, phi = np.meshgrid(theta, phi)

    # Calculamos los valores X, Y, Z
    X = (param1 + param2 * np.cos(phi)) * np.cos(theta)
    Y = (param1 + param2 * np.cos(phi)) * np.sin(theta)
    Z = param2 * np.sin(phi)

    # Devolvemos los valores
    return X, Y, Z

# Creamos una nueva figura
fig = plt.figure(figsize=(12,6))

# Representación Ring torus (param1 > param2)
# Calculamos coordenadas para los parámetros dados
x0, y0, z0 = plot_torus(100, 10, 4)

# Localizamos la gráfica en uno de los subplots
ax0 = fig.add_subplot(2, 2, 1, projection='3d')

# Representamos la gráfica
ax0.plot_surface(x0, y0, z0, alpha=0.9, cmap = cm.viridis)

# Establecemos cómo escalan los ejes y damos título al subplot
ax0.set_aspect('equal')
plt.title('Ring torus')

# Representación Horn torus (param1 = param2)
x1, y1, z1 = plot_torus(100, 4, 4)

ax1 = fig.add_subplot(2, 1, 1, projection='3d')
ax1.plot_surface(x1, y1, z1, alpha=0.9, cmap = cm.viridis)
ax1.set_aspect('equal')
plt.title('Horn torus')

# Representación Spindle torus (param1 < param2)
x2, y2, z2 = plot_torus(100, 4, 10)

ax2 = fig.add_subplot(2, 2, 2, projection='3d')
ax2.plot_surface(x2, y2, z2, alpha=0.9, cmap = cm.viridis)
ax2.set_aspect('equal')
plt.title('Spindle torus')

# Mostramos las gráficas
plt.show()
````

_El root mean square fluctuation (r.m.s.f.) es una medida numérica que mide la flexibilidad individual de diferentes partículas a lo largo de una trayectoria de dinámica molecular. Esta medida normalmente nos sirve para encontrar qué aminoácidos son los que más contribuyen en el movimiento de la proteína._

Las posiciones del vector **datos_dinamica** son las siguientes:

* Número aminoácido
* Valor máximo distancia entre sustrato y aminoácido (Å)
* Valor mínimo distancia entre sustrato y aminoácido (Å)
* Valor medio distancia entre sustrato y aminoácido (Å)
* Valor r.m.s.f. (Å)

````python
#La representación debe devolver gráficos similares a los siguientes
````

````python
%matplotlib inline

from matplotlib import figure
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = 5, 4

datos_dinamica = [["155", 2.394, 1.892, 2.152, 1.478],
                  ["156", 8.198, 4.859, 6.380, 1.389],
                  ["159", 2.327, 1.863, 2.065, 1.535],
                  ["191", 2.501, 1.958, 2.232, 0.815],
                  ["195", 2.358, 1.959, 2.159, 0.959],
                  ["394", 2.305, 1.923, 2.117, 0.768]]

# Respuesta

# Convertimos datos dinámica en array de numpy
np_array = np.array(datos_dinamica)

# Creamos un vector con longitud de los datos
x = np.arange(0, len(datos_dinamica), 1)

# Seleccionamos cada posición del array y asignamos a la variable correspondiente
# Convirtiéndolos en dato tipo float
numero = np_array[:,0]
maximo = np_array[:,1].astype(float)
minimo = np_array[:,2].astype(float)
medio = np_array[:,3].astype(float)
rmsf = np_array[:,4].astype(float)

# Representamos variables maximo, minimo y medio con respecto al nº aminoácido
plt.plot(numero, medio, "-g", label = "Distancia media")
plt.plot(numero, maximo, "-r", label = "Distancia maxima")
plt.plot(numero, minimo, "-b", label = "Distancia minima")

# Colocamos leyenda en la parte superior izquierda
plt.legend(loc="upper right")

# Nombramos los ejes y damos título a la gráfica
plt.xlabel('Número aminoacido', fontsize = 10)
plt.ylabel('Distancia (Å)', fontsize = 10)
plt.title('Distancias substrato - aminoacido')

# Creamos otra figura con un subplot
fig, ax = plt.subplots()

# Representamos el rmsf en forma de gráfico de barras
ax.bar(numero, rmsf, width = 0.8, edgecolor = "white", linewidth = 0.7)

# Nombramos los ejes y damos título a la gráfica
plt.xlabel('Número aminoacido', fontsize = 10)
plt.ylabel('rmsf (Å)', fontsize = 10)
plt.title('rmsf')

# Mostramos los gráficos
plt.show()
````
