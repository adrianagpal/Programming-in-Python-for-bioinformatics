###Ejercicio entrenamiento 1

Añade una linea de código bajo cada comentario para implementar el algoritmo que calculará la distáncia del enlace peptidico (δ(C-N) ≈ 1.32Å) entre dos átomos:

(*Actualizado con tuplas, bucles y funciones*)


image.png

````python
import math

# Definimos las coordenadas de los átomos como puntos 3D mediante tuplas
carbono = (26.100, 29.253, 5.202)
nitrogeno = (26.849, 29.656, 6.217)

# Definimos una función que calcule la distancia entre dos puntos
def calcular_distancia(atom1, atom2):

    # Definimos variable en la que guardar los cuadrados de las diferencias
    calc = 0

    # Por cada item en el rango de la longitud del primer átomo
    for i in range(len(atom1)):

        # Extraemos la misma coordenada de cada átomo y calculamos el cuadrado
        # de la diferencia. Sumamos a la variable 'calc'.
        calc += (atom2[i] - atom1[i]) ** 2

    # Devolvemos el valor final de la variable
    return math.sqrt(calc)

# Aplicamos la función distancia a los dos átomos
print(calcular_distancia(carbono, nitrogeno))
````

###Ejercicio entrenamiento 4

Escribe un programa que dado un texto, cambie todas las `e` o `E` por 3 y `o` u `O` por 0:

````python
texto = "Pero esto importa poco a nuestro cuento: basta que en la narración de él no se salga un punto de la verdad."

# Definimos una función que codifique el texto
def codifica(t):

    # Definimos una variable donde guardar el texto codificado
    texto = ""

    # Por cada letra en nuestro texto
    for letter in t:

        # Si la letra (convertida a minúscula) es igual a "e", transformar en "3"
        # Y adjuntamos a la variable
        if letter.lower() == "e":
            texto += "3"

        # Si la letra (convertida a minúscula) es igual a "o", transformar en "0"
        # Y adjuntamos a la variable
        elif letter.lower() == "o":
            texto += "0"

        # Sino, adjuntamos la letra sin modificaciones a la variable
        else:
            texto += letter

    # Devolvemos el resultado de la variable
    return(texto)

# Imprimimos por pantalla el resultado de la función para nuestro texto
print(codifica(texto))
````

###Ejercicio entrenamiento 5

Escribe un programa que dados dos diccionarios con posibles claves repetidas y valores por cada clave de tipo numérico, devuelva un nuevo diccionario con el contenido de los dos diccionarios originales y, si existe una clave repetida, el valor de esa clave sea la media de los valores originales:

````python
diccionario_1 = {'melón': 1.5, 'sandía': 3.5, 'uva': 1.0, 'mandarina': 3.1}
diccionario_2 = {'melón': 2.5, 'sandía': 1.5, 'platano': 3.0, 'naranja': 4.3}

# Importamos módulo para usar la función de la media
import statistics

# Definimos una función que combine una lista de diccionarios
def media_diccionarios(listaDics):

    # Definimos una lista donde guardar cada clave, valor de los diccionarios
    lista = []

    # Por cada diccionario en la lista de diccionarios
    for item in listaDics:

        # Cogemos cada clave, valor y lo añadimos a la lista como tupla
        lista += list(item.items())

    # Definimos un diccionario donde construir un diccionario combinado
    dict = {}

    # Por cada elemento en la lista con los claves y los valores
    for item in lista:

        # Si el primer elemento de la tupla (clave), no está en el diccionario
        if item[0] not in dict.keys():

            # La añadimos y le damos como valor el segundo elemento de la tupla (valor)
            dict[item[0]] = [item[1]]

        # Si el primer elemento sí está en el diccionario
        else:

            # Adjuntamos el valor al valor existente en forma de lista
            dict[item[0]].append(item[1])

    # Definimos un diccionario donde realizar las medias entre los valores
    new_dict = {}

    # Por cada clave y lista de valores
    for key, value in dict.items():

        # Añadimos la clave al nuevo diccionario, y realimos la media de los
        # valores de la lista
        new_dict[key] = statistics.mean(value)

    # Devolvemos el nuevo diccionario
    return new_dict

# Imprimimos por pantalla el diccionario combinado
print(media_diccionarios([diccionario_1, diccionario_2]))
````

````python
# Masas atómicas
masas = {'H': 1.007825, 'C': 12.01, 'O': 15.9994,
         'N': 14.0067, 'S': 31.972071, 'P': 30.973762}

# Definimos una función que calcule la masa atómica de una molécula
def calcula_masa_atomica(molecula):
    """Calcula la masa atomica de una molecula"""

    # Definimos una variable donde acumular las masas parciales
    masa = 0.0

    # Separamos molécula por '-'
    grupos = molecula.split('-')

    # Recorremos grupos de la molécula
    for atomo in grupos:

        # Extraemos elemento y número de átomos por cada grupo
        elemento = "".join(ch for ch in atomo if ch.isalpha())
        numero = "".join(ch for ch in atomo if ch.isnumeric())

        # Si hay más de un átomo del mismo tipo, multiplicamos el número de átomos
        # por la masa (obtenida del diccionario)
        if len(numero) != 0:
            masa += int(numero) * masas[elemento]

        # Si sólo hay un átomo de su tipo, sumamos su masa una vez
        else:
            masa += masas[elemento]

    # Devolvemos la masa total con 4 decimales
    return round(masa, 4)
````

### Ejercicio 5 **<font color="blue" size="+2">EG</font>**

Uno de los algoritmos más básicos en criptografía es el cifrado César (https://es.wikipedia.org/wiki/Cifrado_C%C3%A9sar), que fue utilizado por Julio César para comunicarse con sus generales, y que consiste en dado un texto, por cada una de las letras del texto, añadirle un desplazamiento para conseguir una nueva letra diferente de la original. Comprenderemos rápidamente su mecanismo mediante un ejemplo:

Si asignamos el número 1 a la primera letra del abecedario, A, 2 a la siguiente, B, etc., imaginad que tenemos el siguiente mensaje:
<code>
ABC
123
</code>

Si aplicamos un desplazamiento de 3, buscaremos cuál es la letra en el abecedario que se corresponde:
<code>
DEF
456
</code>

ABC se ha convertido en DEF porque hemos sumado un desplazamiento de 3. También podríamos aplicar otros tipos de desplazamiento como los negativos. Por ejemplo, para el desplazamiento -1 y el mensaje original ABC tendríamos un mensaje cifrado de: ZAB.

Escribid una función que **dado un mensaje original y un desplazamiento, calcule y devuelva el mensaje cifrado**:

````python
"""
Podemos simplificar sin problemas el uso de palabras solo
a mayúsculas y al abecedario en inglés.
Vamos a utilizar la propiedad de que la primera letra del
abecedario A se corresponde con el entero
65 en ASCII y Z con el 90, por lo tanto, tenemos una longitud
de abecedario de 90-65+1=26 letras.
"""

# Definimos una función que cifre el mensaje con un dado desplazamiento
def cifrado_cesar(mensaje, desplazamiento=1):

    # Creamos una variable para guardar el mensaje cifrado
    mensaje_cifrado = ""

    # Cogemos el valor numérico de la primera letra de nuestro abecedario
    inicial = ord("A")

    # Por cada elemento en el mensaje
    for letra in mensaje:

        # Comprobamos si el elemento es una letra
        if letra.isalpha():

            # Cogemos el valor numérico en ASCII de la letra
            indice = ord(letra)

            # Sumamos el valor con el desplazamiento y restamos el valor inicial.
            # La operación módulo se realiza para que el desplazamiento
            # o el nuevo índice se encuentre entre 0 y 25.
            nuevo_indice = (indice - inicial + (desplazamiento % 26))% 26

            # Obtenemos la nueva letra de acuerdo al nuevo índice
            nueva_letra = chr(inicial + nuevo_indice)

            # Incluimos la nueva letra en nuestra variable
            mensaje_cifrado += nueva_letra

        else:
            # Si el elemento no es una letra, simplemente lo incluimos en la variable
            mensaje_cifrado += letra

    # Devolvemos el valor del mensaje cifrado
    return mensaje_cifrado
````

````python
%%ipytest
# Test Ejercicio 5
def test_cifrado_cesar():
  assert cifrado_cesar('ABCD', 27) == "BCDE"
  assert cifrado_cesar('ABCD', 1) == "BCDE"
  assert cifrado_cesar('XYZ', 1) == "YZA"
  assert cifrado_cesar("PROGRAMACION PARA LA BIOINFORMATICA", 26) == "PROGRAMACION PARA LA BIOINFORMATICA"
  assert cifrado_cesar("PROGRAMACION PARA LA BIOINFORMATICA", 2) == "RTQITCOCEKQP RCTC NC DKQKPHQTOCVKEC"
````

````python
numberList = [10, 20, 30, 40, 10]

# Definimos una función para comprobar si el primer y último número son iguales
def first_last_same(numberList):

    # Cogemos el primer elemento de la lista
    primero = numberList[0]

    # Cogemos el último elemento de la lista
    ultimo = numberList[-1]

    # Comprobamos si el primer y último valor son iguales, y devolvemos booleano
    return primero == ultimo

# Imprimimos el resultado por pantalla usando f-strings
print(f"Dada la lista: {numberList} \nEl primer y el último número de la lista son iguales {first_last_same(numberList)}")
````

### Ejercicio 9 **<font color="green" size="+2">MU</font>**

Dadas dos listas de números, cread una nueva lista que contenga los números **impares** de las dos listas

````python
# Definimos una función para crear una lista con impares
def merge_list(list1, list2):

    # Definimos una variable donde guardar la lista de resultados
    result_list = []

    # Combinamos las listas
    lista_conjunta = list1 + list2

    # Por cada elemento en la lista conjunta
    for item in lista_conjunta:

        # Si el número de la lista no es divisible entre 2 (es impar),
        # lo añadimos a la lista de resultados
        if item % 2 != 0:
            result_list.append(item)

    # Devolvemos la lista de resultados
    return result_list
````

### Ejercicio 10 **<font color="green" size="+2">MU</font>**

Escribe una función llamada `distance` que calcule la distancia entre dos puntos en el espacio 3D. Estos puntos son (x1, y1, z1) y (x2, y2, z2):

````python
import math

# Definimos una función que calcule la distancia entre 2 puntos dadas sus coordenadas
def distance(x1, y1, z1, x2, y2, z2):

    """Calcula la distancia euclidiana entre dos puntos 3D"""

    # Calculamos la suma de los cuadrados de las diferencias
    suma_coord_cuad = (x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2

    # Calculamos la raíz cuadrada
    dist = math.sqrt(suma_coord_cuad)

    # Devolvemos el resultado de la variable
    return dist
````

````python
%%ipytest
import pytest
# Test Ejercicio 10
def test_distance():
  assert distance(1.0, 9.0, 1.0, 2.0, 4.0, 6.0) == pytest.approx(7.14142842854285)
  assert distance(3.0, 2.0, -6.0, 7.0, -3.0, 1.0) == pytest.approx(9.486832980505138)
  assert distance(5.0, 1.0, 6.0, -2.0, 15.0, 8.0) == pytest.approx(15.7797338380595)
````

El paradigma de programación orientada a objetos es ampliamente utilizado en gran parte de las librerías que se escriben en Python. Es una forma útil de encapsular información de la que se ocupará el propio objeto donde se ha definido esa información.

**Explicad qué son los siguientes conceptos:**

* Una clase
* Un objeto
* Un atributo
* Un método
* Un constructor
* Una superclase y una subclase

**Poned un ejemplo de definición de una clase en código Python y un ejemplo de uso de esa misma clase**.


**Respuesta:**

- Clase. Estructura de datos que actúa como plantilla para definir un conjunto de atributos y métodos, que describen un objeto.

- Objeto. Es una entidad dentro de un lenguaje de programación que nos sirve para representar conceptos o cosas en el contexto del programa. Este objeto contiene un estado, definido por los valores de sus atributos, y un comportamiento, dado por los métodos que nos permiten interactuar con el objeto.

- Atributo. Es una variable compartida por todas las instancias u objetos de una clase, que almacena datos que pertenecen al objeto de esa clase.

- Método. Es una función perteneciente a un objeto y que nos permite interaccionar con él, modificando los valores de los atributos del objeto.

- Constructor. Es una función o un método que sirve para inicializar un objeto y asignarle los atributos que necesite.

- Superclase y subclase. Una subclase es una clase que hereda los atributos y métodos de otra clase, llamada superclase.

````python
# Creamos una clase
class Folder:

    # Inicializamos el objeto y le damos un nombre (atributo)
    def __init__(self, name):
        self.data = []
        self.name = name

    # Método para devolver información sobre el objeto en forma de string
    def __str__(self):
        return f'El objeto {self.name} tiene los siguientes ficheros: {self.data}.'

    # Método para añadir ficheros al objeto
    def add(self, file):
        self.data.append(file)

# Creamos un objeto de la clase 'Folder'
carpeta = Folder("Mis Documentos")

# Añadimos un fichero al objeto
carpeta.add("Ejercicios.txt")

# Imprimimos por pantalla la información sobre el objeto
print(carpeta)
````

Las excepciones son errores detectados en tiempo de ejecución. Pueden y deben ser manejadas por el programador para minimizar el riesgo de que un determinado programa falle de forma no controlada.

**Poned ejemplos de tipos de excepción en el lenguaje Python y de cómo se capturan.**

**Respuesta:**

Algunos ejemplos de excepciones en Python son los siguientes:

*   SyntaxError. Hace referencia a errores de sintaxis.
*   TypeError. Se produce si intentamos realizar operaciones que no soportan el uso de dos tipos (por ejemplo, string y un entero).
*   KeyError. Se produce cuando no se encuentra una clave en un diccionario.
*   IndexError. Se produce cuando llamamos a un elemento fuera de un rango.

Podemos capturar estas excepciones con *try* y *except*.

````python
# Ejemplo para capturar excepción TypeError

a = "hola"
b = 5

try:
    texto = a + b
    print(texto)
except TypeError:
    print("La operación no soporta los tipos de datos.")

# Ejemplo para capturar excepción KeyError

dict = {"Naranjas": 2, "Fresas": 3, "Albaricoques": 5}
fruta = "Plátanos"

try:
    print(dict[fruta])
except KeyError:
    print("La clave no se encuentra en el diccionario.")
````
